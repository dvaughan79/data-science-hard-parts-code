```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# set plotting parameters from the beginning
font = {'family' : 'monospace',
        'weight' : 'normal',
        'size'   : 14}
axes = {
        'titlesize' : 22,
        'labelsize' : 20}
matplotlib.rc('font', **font)  # pass in the font dict as kwargs
matplotlib.rc('axes', **axes)  # pass in the font dict as kwargs
```

##### Let's simulate data for the loan origination example

* Simulate two positively correlated normal variables

```python
# simulate two positively correlated random variables
N = 100
mean = [0.5, 0.5]
cov = [[1,0.8],[0.8,1]]
col_labels = ['default_prob', 'accept_prob']
data = pd.DataFrame(np.random.multivariate_normal(mean, cov, size=N), columns = col_labels)
# normalize so that they look like probability scores
min_data = data.min()
max_data = data.max()
data_norm = (data - min_data)/(max_data - min_data)
```

```python
# plot
fig, ax = plt.subplots(figsize =(10,6))
ax.scatter(data_norm['default_prob'], data_norm['accept_prob'])
# get plot thresholds
mean_x1 = data_norm['default_prob'].mean()
mean_x2 = data_norm['accept_prob'].mean()
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
ax.plot([xmin, xmax], [mean_x2, mean_x2], color = 'k', ls = '--', alpha = 0.5)
ax.plot([mean_x1, mean_x1], [ymin, ymax], color = 'k', ls = '--', alpha = 0.5)
ax.axis([xmin, xmax, ymin, ymax])
ax.set_title('2x2 for Credit Origination')
ax.set_ylabel('Accept Probability')
ax.set_xlabel('Default Probability')
# get some labels
ax.text(0.93, 0.95, 'A', horizontalalignment='center',verticalalignment='center',
        weight = 'bold', fontsize = 20, transform = ax.transAxes)
ax.text(0.03, 0.95, 'B', horizontalalignment='center',verticalalignment='center',
        weight = 'bold', fontsize = 20, transform = ax.transAxes)
ax.text(0.03, 0.45, 'C', horizontalalignment='center',verticalalignment='center',
        weight = 'bold', fontsize = 20, transform = ax.transAxes)
ax.text(0.93, 0.45, 'D', horizontalalignment='center',verticalalignment='center',
        weight = 'bold', fontsize = 20, transform = ax.transAxes)
```
